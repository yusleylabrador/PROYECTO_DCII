**Importancia de la Salud Mental en los Estudiantes**
La salud mental de los estudiantes es un tema de creciente importancia en el ámbito educativo y social. La Organización Mundial de la Salud (OMS) define la salud mental como un estado de bienestar en el que una persona puede hacer frente a los momentos de estrés de la vida, desarrollar todas sus habilidades, aprender y trabajar adecuadamente, y contribuir a la mejora de su comunidad. En este contexto, la salud mental de los estudiantes es fundamental para su éxito académico y personal.

Varios factores pueden afectar la salud mental de los estudiantes, incluyendo la presión académica, la presión laboral, el estrés financiero, y el historial familiar de enfermedades mentales. Además, los hábitos de sueño y alimentación, así como la satisfacción con los estudios y el trabajo, también juegan un papel crucial en el bienestar mental de los estudiantes.

Estudios recientes han destacado la creciente prevalencia de problemas de salud mental entre los estudiantes universitarios. La pandemia de COVID-19 ha exacerbado estos problemas, introduciendo nuevos desafíos como el aislamiento social y la transición a la educación en línea. Un estudio realizado por la Universidad Nacional de Avellaneda (UNDAV) encontró que la presión académica, la adaptación a nuevos entornos sociales y la gestión del tiempo son factores significativos que contribuyen al estrés y la ansiedad en los estudiantes.

Otro estudio revisó la influencia de la salud mental en el desempeño académico durante la pandemia y encontró que la incorporación de estrategias de salud mental en el aprendizaje puede mejorar significativamente el rendimiento académico de los estudiantes. Además, se ha demostrado que las intervenciones para la salud mental, como la psicoeducación y el apoyo psicológico, son efectivas para mejorar el bienestar emocional de los estudiantes.

En nuestro conjunto de datos, tenemos algunas columnas con datos categóricos. Pero para el proceso de aprendizaje automático, los transformamos a numericos


```python
def convertir_num(df, col):
    # Creamos un diccionario
    unicos = df[col].unique()
    valor_a_numero = {value: idx + 1 for idx, value in enumerate(unicos)}

    # Convertimos en columna
    df[col] = df[col].map(valor_a_numero)

    return df, valor_a_numero
```

```python
# Crear una tabla dinámica para el conteo de depresión por género
pivot_depression_gender = df.pivot_table(values='Depression', index='Gender', aggfunc='sum').reset_index()
pivot_depression_gender.columns = ['Gender', 'Count_Depression']

# Crear una tabla dinámica para el conteo de pensamientos suicidas por género
pivot_suicidal_gender = df.pivot_table(values='Have you ever had suicidal thoughts ?', index='Gender', aggfunc='sum').reset_index()
pivot_suicidal_gender.columns = ['Gender', 'Count_Suicidal_Thoughts']

# Combinar las tablas dinámicas utilizando merge
merged_pivot = pd.merge(pivot_depression_gender, pivot_suicidal_gender, on='Gender')

# Mostrar la tabla combinada
print(merged_pivot)

```

Vemos en los diferentes graficos que el género masculino han sido quienes han tenido mayores pensamientos suicidas, vemos que estos estan sometidos a más horas de estudio, mayor estres financiero, han padecido en mayor frecuencia de depresión y mayor presion académica.
